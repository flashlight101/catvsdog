Задание 1
Два стрелка по очереди стреляют по мишени до первого попадания. Случайная
величина X – количество выстрелов, сделанных первым стрелком. Случайная
величина Y – количество выстрелов, сделанных вторым стрелком. Построить ряды
распределения случайных величин X, Y, X+Y и XY, если известно, что у каждого из
них имеется по 5 патронов, а вероятность попадания при одном выстреле у первого
стрелка равна 0,6, а у второго – 0,4

```python
import numpy as np
import pandas as pd
```

```python
X=np.array([0.01024,0.6,0.24,0.096,0.0384,0.01536])#Ряд распределения X
Y=np.array([0.0776,0.4,0.24,0.144,0.864,0.05184])#Ряд распределения Y
```

```python
data=np.dot(Y.reshape(6,1), X.reshape(1,6))
data=pd.DataFrame(data)
row=np.empty([11,1])
for i in range(6):
    for j in range(6):
        row[i+j]+=data[i][j]
row=pd.DataFrame(row)
row[0] = row[0].apply(lambda x: '{:.10f}'.format(x))
```

X+Y

```python
row
```

```python
data=np.dot(Y.reshape(6,1), X.reshape(1,6))
data=pd.DataFrame(data)
row=np.empty([26,1])
for i in range(6):
    for j in range(6):
        row[i*j]+=data[i][j]
row=pd.DataFrame(row)
row[0] = row[0].apply(lambda x: '{:.10f}'.format(x))
row = row.loc[row[0] !='0.0000000000' ]
```
